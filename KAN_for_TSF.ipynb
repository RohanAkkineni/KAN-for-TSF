```python
# Create the CNN model
def create_cnn_model(input_shape):
    model = keras.Sequential()
    model.add(layers.Conv1D(64, kernel_size=2, activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling1D(pool_size=2))
    model.add(layers.Conv1D(32, kernel_size=2, activation='relu'))
    model.add(layers.MaxPooling1D(pool_size=2))
    model.add(layers.Flatten())
    model.add(layers.Dense(50, activation='relu'))
    model.add(layers.Dense(1))  # Output layer
    return model

# Reshape for CNN, keeping it 2D, treating it like a standard regression problem
X_train_reshaped = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))  # 3D for Conv1D
X_test_reshaped = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Create and compile the model
cnn_model = create_cnn_model((X_train_reshaped.shape[1], 1))
cnn_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
cnn_model.fit(X_train_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)

# Make predictions
predictions = cnn_model.predict(X_test_reshaped)

# Evaluate the model
cnn_mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {cnn_mse}')
```
